```python
import os
from elasticsearch import Elasticsearch, helpers, NotFoundError
import json
from datetime import datetime
```

```python
elastic_user = "elastic"
elastic_pwd = "youm6sP3"
```

```python
client = Elasticsearch("http://localhost:9200", \
                       basic_auth=(elastic_user, elastic_pwd))  
```

```python
#q1
response = client.info()
with open('answers/q1.json', 'w') as f:
    json.dump(dict(response), f, indent=4)
# response
```

```python
# q2

name = 'madmap'

json_data_dir = "data/jsons"
# json_files = [f for f in os.listdir(json_data_dir) if ".json" in f]
# json_files
operations = []

for news in os.listdir(json_data_dir):
    if news.endswith('.json') and news != 'places.json':
        with open(os.path.join(json_data_dir, news), 'r') as f:
            data = json.load(f)
            operations.append({
                "index": {"_index":name}
            })
            operations.append(data)

if operations : 
    bulk_output = client.bulk(operations = operations)
response = client.indices.get_mapping(index=name)
            
with open('answers/q2.json', 'w') as f:
    json.dump(dict(response), f, indent=4)
# response
```

```python
#q3
text_data = 'data/text'
operations = []
name = 'madmap'
mapping_update = {
    "properties": {
        "wiki": {
            "type": "text"
        }
    }
}
client.indices.put_mapping(index=name, body=mapping_update)
for text in os.listdir(text_data):
    with open(os.path.join(text_data, text), 'r') as f:
        loaded_data = f.read()
        operations.append({
            'index': {'_index':name}
        })
        operations.append({'wiki': loaded_data})
bulk_output = client.bulk(body=operations)

response = client.indices.get_mapping(index=name)
with open('answers/q3.json', 'w') as f:
    json.dump(dict(response), f, indent=4)
# response
```

```python
#q4

response = client.search(
    index='madmap', 
    body ={
        'query': {
            'simple_query_string': {
                'query': "rivalry^3 \Badgers^2 football",
                'fields': ['wiki']
            }
        },
        'size':3
    }
)
with open('answers/q4.json', 'w') as f:
    json.dump(dict(response), f, indent=4)

# for hit in response['hits']['hits']:
#     print(f"Score: {hit['_score']}, Source: {hit['_source']}")
response
```
